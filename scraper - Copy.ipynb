```python
import time
from facebook_scraper import *

results = []
pagination_url = None

def handle_pagination_url(url):
    global pagination_url
    pagination_url = url
    print("Pagination URL:", pagination_url)  # Print the pagination URL

set_cookies("cookies.txt")

while True:
    try:
        for post in get_posts("RevenueDepartmentOdisha", page_limit=None, request_url_callback=handle_pagination_url):
            print(len(results))
            print(post['text'][:100])
            print(post['time'])
            results.append(post)

        # Check if there's a pagination URL, and break the loop if there isn't one
        if pagination_url is None:
            print("No more pages to scrape.")
            break

        # Use pagination_url to continue scraping the next page
        for post in get_posts("RevenueDepartmentOdisha", page_limit=None, start_url=pagination_url, request_url_callback=handle_pagination_url):
            print(len(results))
            print(post['text'][:100])
            print(post['time'])
            results.append(post)

        print("All done")
        break
    except exceptions.TemporarilyBanned:
        print("Temporarily banned, sleeping for 10m")
        time.sleep(600)
```
